```python
import matplotlib.pyplot as plt
import pandas as pd
ucl = pd.read_csv('/home/neji/Downloads/DATA/defending.csv')
recovered = ucl.sort_values("balls_recoverd", ascending=False).head(10)
recovered.plot(x="player_name", y="balls_recoverd", kind="bar", color="lightcoral")
plt.xlabel("Player Name", fontweight="bold", fontsize=12)
plt.ylabel("Number of Balls Recovered", fontweight="bold", fontsize=12)
plt.yticks(fontsize=10, fontweight="bold")
plt.xticks(fontsize=10, fontweight="bold")
plt.title("Number of Balls Recovered by a Player in UCL 2021/22 (TOP 10)", fontweight="bold")
plt.legend(["Balls Recoverd"])
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
ucl = pd.read_csv('/home/neji/Downloads/DATA/defending.csv')
t_won = ucl.sort_values("t_won", ascending=False).head(10)
t_won.plot(x="player_name", y="t_won", color="palevioletred", kind="bar")
plt.ylabel("NUMBER OF TACKLES WON", fontweight="bold", fontsize=12)
plt.xlabel("PLAYER NAME", fontweight="bold", fontsize=12)
plt.yticks([0, 4, 8, 12, 16])
plt.xticks(fontsize= 10, fontweight="bold")
plt.yticks(fontsize= 10, fontweight="bold")
plt.title("Number of Tackles Won by a player in UCL 2021/22 (TOP 10)", fontweight="bold")
plt.legend(["Tackles Won"])
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
ucl = pd.read_csv('/home/neji/Downloads/DATA/defending.csv')
clearance = ucl.sort_values('clearance_attempted', ascending=False).head(10)
clearance.plot(x="player_name", y="clearance_attempted", kind="bar", color="lightseagreen")
plt.xlabel("PLAYER NAME", fontweight="bold", fontsize=12)
plt.ylabel("CLEARANCE ATTEMPTED", fontweight="bold", fontsize=12)
plt.xticks(fontsize= 10, fontweight="bold")
plt.yticks(fontsize= 10, fontweight="bold")
plt.title("CLEARANCE ATTEMPTED BY A PLAYER IN UCL 2021/22 (TOP 10)", fontweight='bold')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
ucl = pd.read_csv('/home/neji/Downloads/DATA/defending.csv')
ucl["avg_t_won_perMatch"] = ucl["t_won"] / ucl["match_played"]
avg_t_won_perMatch = ucl.sort_values("avg_t_won_perMatch", ascending=False).head(10)
avg_t_won_perMatch.plot(x="player_name", y="avg_t_won_perMatch", kind="bar", color='royalblue')
plt.title("Average Tackle won by a player per Match", fontweight="bold")
plt.xticks(fontsize= 10, fontweight="bold")
plt.yticks(fontsize= 10, fontweight="bold")
plt.ylabel("Average", fontweight='bold', fontsize=10)
plt.xlabel("Player Name", fontweight='bold', fontsize=12)
plt.legend(['Average'])
plt.show()
```
