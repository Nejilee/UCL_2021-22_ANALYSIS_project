```python
import matplotlib.pyplot as plt
import pandas as pd
ucl = pd.read_csv('/home/neji/Downloads/DATA/defending.csv')
clearance = ucl.sort_values("balls_recoverd", ascending=False).head(10)
clearance.plot(x="player_name", y="balls_recoverd", kind="bar", color="limegreen")
plt.xlabel("Player Name", fontweight="bold", fontsize=10)
plt.ylabel("Number of Ball Clearances", fontweight="bold", fontsize=10)
plt.yticks(fontsize=10, fontweight="bold")
plt.xticks(fontsize=10, fontweight="bold")
plt.title("Highest Number of Ball Clearances by a Player in UCL 2021/22", fontweight="bold")
plt.legend(["Balls Recoverd"])
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
ucl = pd.read_csv('/home/neji/Downloads/DATA/defending.csv')
t_won = ucl.sort_values("t_won", ascending=False).head(10)
t_won.plot(x="player_name", y="t_won", color="palevioletred", kind="bar")
plt.ylabel("Number of Tackles Won ", fontweight="bold")
plt.xlabel("Player Name", fontweight="bold")
plt.yticks([0, 4, 8, 12, 16])
plt.xticks(fontsize= 10, fontweight="bold")
plt.yticks(fontsize= 10, fontweight="bold")
plt.title("Highest Number of Tackles Won by a player in UCL 2021/22", fontweight="bold")
plt.legend(["Tackles Won"])
plt.show()
```
